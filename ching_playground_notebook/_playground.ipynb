```python
pres_results = pandas.read_csv("./data4democracy-election-transparency/PresidentialElectionResults2016.csv")
```

```python
pres_results
```

```python
al_results = pres_results[pres_results['StateName'] == 'Alabama']
al_results
```

```python
f,axes = plt.subplots(1,2)
axes[0].hist(al_results['clinton'])
axes[1].hist(al_results['trump'])
plt.show()
```

```python

dem_results = pres_results[pres_results['dPct'] > .5]
dem_results
```

```python
f,axes = plt.subplots(1,2)
axes[0].hist(dem_results['clinton'])
axes[1].hist(dem_results['trump'])
plt.show()
```

```python
small_counties = pres_results[pres_results['totalvotes'] <= 20000]
small_counties
```

```python
f,axes = plt.subplots(1,2)
axes[0].hist(small_counties['clinton'])
axes[1].hist(small_counties['trump'])
plt.show()
```

```python
f,axes = plt.subplots(1,2)
axes[0].scatter(small_counties['dPct'], small_counties['leanD'], alpha=.1)
axes[1].scatter(small_counties['rPct'], small_counties['leanR'], alpha=.1)
plt.show()
```
